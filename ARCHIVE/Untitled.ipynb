```python
import pandas as pd
import scipy.stats
import numpy as np
import gspread
```

```python
gc = gspread.oauth()

sh = gc.open_by_url('https://docs.google.com/spreadsheets/d/1KlVfVCuob9hinhDu_Pvk6Z1kctlbnA2Ct14gnpVcx6U/edit#gid=0')

ws = sh.worksheet('Data')

df = pd.DataFrame(ws.get_all_records())
df.head()
```

```python
r2_df_dict = {}

for col in df.columns[3:]:
    correlation_matrix = np.corrcoef(df['wins'],col)
    correlation_xy = correlation_matrix[0,1]
    r2 = correlation_xy**2
    print(col + ' ' + ' ' + str(r2))
    if col not in r2_df_dict.keys():
        r2_df_dict[col] = r2
```

```python
from sklearn.metrics import r2_score

r2_score(df[7], df[8])
```
